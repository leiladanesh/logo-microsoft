```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.full((400,800,3),255,dtype="uint8")
```

```python
img[0:400,0:800]=[120,120,120]
img[130:200,130:200]=[255,0,0]
img[210:280,210:280]=[255,165,0]
img[210:280,130:200]=[0,0,255]
img[130:200,210:280]=[0,255,0]

```

```python
image=cv2.putText(img,'Microsoft',(300,220),cv2.FONT_ITALIC,3,(255,255,255),8)

plt.imshow(image)
```
